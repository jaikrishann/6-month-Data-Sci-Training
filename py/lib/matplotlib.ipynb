```python
x= [1,2,3,4,5]
y = [1,10,9,40,25]
plt.plot(x,y,marker = "D",ms = 10,mfc = "red",mec="black",color = "green",lw = 3,ls = "-.",label = "line plot",alpha = 0.9)
plt.title("line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()



plt.show()
```

```python
##scatter plot 
plt.scatter(x,y,marker = "D",s = 100,c = "red",alpha = 0.9,label = "scatter plot",edgecolor = "black",linewidth = 3)
```

```python
##bar chart 
x = ["mp","rj","up","mb"]
y = [1000,4000,8000,12000]
colours = ["red","green","blue","yellow"]
hatch_patterns = ['/', '\\', '|', '-'] 

plt.bar(x,y,color = colours,label = x,width = 0.5,edgecolor = "black",linewidth = 3,alpha = 0.9,hatch = hatch_patterns,zorder = 3)
plt.legend()
plt.grid()
```

```python
plt.barh(x,y,color = colours)
```

```python
##histogram -- distribution of data in bins(intervals)
data = np.random.randn(1000)

plt.hist(data, bins=30, color='skyblue', edgecolor='black',hatch='//')
plt.show()
```

```python
##piechart 
labels = ['A', 'B', 'C', 'D']
sizes = [30, 15, 45, 10]
explodes = (0.1,0,0,0.1)

plt.pie(sizes, labels=labels, autopct='%1.2f%%', colors=['red', 'blue', 'green', 'yellow'],shadow=True,explode=explodes)
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, np.sin(x))
plt.title("Sin(x)")

plt.subplot(2, 2, 2)  # 2nd subplot
plt.plot(x, np.cos(x))
plt.title("Cos(x)")

plt.subplot(2, 2, 3)  # 3rd subplot
plt.plot(x, np.tan(x))
plt.title("Tan(x)")

plt.subplot(2, 2, 4)  # 4th subplot
plt.plot(x, np.exp(-x))
plt.title("Exp(-x)")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
