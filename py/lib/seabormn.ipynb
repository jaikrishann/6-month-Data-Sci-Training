```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
plt.plot(df["total_bill"])
```

```python
sns.lineplot(df["total_bill"])
```

```python
# sns.scatterplot(df["total_bill"])
sns.scatterplot(x="total_bill", y="tip",data=df)
```

```python
plt.title("Total Bill vs Tip",fontsize = 30,color="red",backgroundcolor="black",weight="bold",style="italic",loc="center",)
sns.histplot(df["total_bill"],color="red",edgecolor="red")
```

```python
sns.barplot(x="day",y="total_bill",data=df)
```

```python
sns.kdeplot(df["total_bill"],fill= True)
```

```python
##catogrical --catplot 
# numerical //relational --relplot
```

```python
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",hue = "smoker",style="size",col="day")
```

```python

sns.catplot(x="day",y="total_bill",data=df,kind="bar",hue="smoker",col="sex")
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="swarm")
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="box")
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="violin",)
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="point",aspect=2)
```
