```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("final_df.csv")
df.head()
```

```python
x= df.drop(["Customer Type_disloyal Customer"], axis=1)
y = df[["Customer Type_disloyal Customer"]]
```

```python
# type(df["Customer Type_disloyal Customer"])
df[["Customer Type_disloyal Customer"]]
```

```python
y 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
##decision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
##plotting the decision tree
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dt, filled=True, feature_names=x.columns)
plt.show()
```
